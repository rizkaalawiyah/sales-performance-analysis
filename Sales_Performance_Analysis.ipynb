SELECT o.Date as order_date <br>
pc.CategoryName as category_name,<br>
        p.ProdName as product_name, <br>
        p.Price as product_price, <br>
        o.Quantity as order_qty, <br>
        ROUND(SUM(p.Price * o.Quantity), 2) as total_sales, <br>
        c.CustomerEmail as cust_email, <br>
        c.CustomerCity as cust_city <br>
FROM Final_Task.Orders o <br>
JOIN Final_Task.Customers c ON o.CustomerID=c.CustomerID <br>
JOIN Final_Task.Products p ON o.ProdNumber=p.ProdNumber <br>
JOIN Final_Task.ProductCategory pc ON p.Category=pc.CategoryID <br>
GROUP BY order_date, category_name, product_name, product_price, order_qty, cust_email, cust_city <br>
ORDER BY order_date ASC; <b

Bberdasarkan hasil analisis penjualan yang dilakukan pada PT Sejahtera Bersama tahun 2020-2021 diketahui bahwa ada penurunan performa penjualan pada tahun 2021 dari 2020. Berikut ini merupakan beberapa rekomendasi yang dapat dilakukan agar penjualan optimal:

1. Memperluas target pasar
Asal kota pelanggan menunjukan perbedaan signifikan dalam kontribusi pelanggan, dengan beberapa wilayah memiliki sedikit atau bahkan tidak ada pelanggan, sementara wilayah lainnya memiliki pelanggan yang signifikan. Hal ini menunjukan kebutuhan untuk mengoptimalkan strategi promosi agar dapat meratakan cakupan pasar di seluruh wilayah. Dengan demikian, diperlukan langkah-langkah untuk memastikan bahwa produk lebih dikenal dan dapat menjangkau calon konsumen di semua wilayah.

2. Feedback customer untuk mengetahui tren
Berdasarkan hasil analisis terdapat perbedaan jauh penjualan produk dari tahun sebelumnya untuk itu mendapatkan feedback pelanggan secara aktif perlu dilakukan untuk memahami alasan di balik preferensi customer terhadap produk dan untuk meningkatkan produk dan layanan.

3. Melakukan paket penjualan (bundling)
berdasarkan analisis customer cenderung membeli produk hanya pada kategori tertentu untuk melakukan optimasi penjualan pada produk lain dapat dilakukan paket penjualan (bundling). Hal ini dapat mendorong pelanggan untuk mencoba produk-produk lain.
